```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import linear_model
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import seaborn as sns
```

```python
data_untested_molec =pd.read_csv('tested_molecules_with_descriptors.csv')
data_untested_molec.head()

df_descriptors = data_untested_molec[['PKM2_inhibition', 'ERK2_inhibition']]
df_data= data_untested_molec.drop(columns=['PKM2_inhibition', 'ERK2_inhibition', 'SMILES'])
df_descriptors.head()

zero_variance_columns = df_data.columns[df_data.nunique() <= 1]
# remove columns of zero variance
df_data = df_data.drop(columns=zero_variance_columns)

scalerminmax = MinMaxScaler()
df_scaled = scalerminmax.fit_transform(df_data)


pca = PCA() 
df_pca=pca.fit_transform(df_scaled)
cumulative_explained_variance = pca.explained_variance_ratio_.cumsum()

# Determine the number of components needed to explain 90% of the variance
desired_variance = 0.90
num_components = (np.argmax(cumulative_explained_variance >= desired_variance) + 1)
#loose components who have too little influence
pca=PCA(n_components=num_components)

df_untested_mol_pca = pca.fit_transform(df_pca)
df_pca = pd.DataFrame(df_untested_mol_pca, columns=[f'PC{i+1}' for i in range(num_components)])

df_pca.dropna(inplace=True)

X_train, X_test, y_train, y_test = train_test_split(df_pca,df_descriptors, test_size=0.2)

for descriptor in range(X_train.shape[1]):
    X_train_feature= X_train.iloc[:, descriptor].values.reshape(-1, 1)
    X_test_feature = X_test.iloc[:, descriptor].values.reshape(-1, 1)


    # Create linear regression object
    model = LinearRegression()

    # Fit the model using the training sets
    model.fit(X_train_feature, y_train)

    # Make predictions using the testing set
    y_pred = model.predict(X_test_feature)
    # Evaluation metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred) 
    mse = mean_squared_error(y_test, y_pred)
    print(f"For descriptor{descriptor}:")
    print(f"R^2 score: {r2}")
    print(f"MAE: {mae:.2f}")
    print(f"Mean Squared Error: {mse}")
    print(len(X_test_feature), len(y_test))
    # Plot outputs
    plt.figure()
    #plt.scatter(X_test_feature, y_test, color="black")
    plt.plot(X_test_feature, y_pred, color="blue", linewidth=3)

    plt.xlabel(f"descriptor {descriptor}")
    plt.ylabel("Target")
    plt.title(f"Linear Regression: descriptor {descriptor} vs Target")
    plt.show()




```
