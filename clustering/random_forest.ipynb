```python
import pandas as pd
```

```python
# reading the data: 
df_pca = pd.read_csv(r"C:\Users\20223319\OneDrive - TU Eindhoven\Bestanden\Jaar 2\Q4\Advanced Programming\Group Assignment\PCA_data.csv")
df_pca.head(5)
```

```python
from sklearn.model_selection import train_test_split

# splitting the data into test and training data:
df_X = df_pca.drop(["PKM2_inhibition","ERK2_inhibition"],axis=1)
df_Y = df_pca[["PKM2_inhibition","ERK2_inhibition"]]

X_train, X_test, y_train, y_test = train_test_split(df_X, df_Y, train_size=0.8)  
```

```python
from sklearn.ensemble import RandomForestClassifier

# fitting data:
classifier = RandomForestClassifier(100)
classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report

# predicting data and testing accuracy:
predict = classifier.predict(X_test)
score = classifier.score(X_test,y_test)
score
```
